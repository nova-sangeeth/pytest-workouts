For example, 2 is written as II in Roman numeral, just two one's added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9. 
X can be placed before L (50) and C (100) to make 40 and 90. 
C can be placed before D (500) and M (1000) to make 400 and 900.

```python
def romanToInt(s: str) -> int:
    romans = {
        'I': 1,
        'V': 5,
        'X': 10,
        'L': 50,
        'C': 100,
        'D': 500,
        'M': 1000
        }
    val = 0
    while(len(s)):
        if len(s) > 1 and (
            # I can be placed before V (5) and X (10) to make 4 and 9. 
            s[0] == 'I' and s[1] == 'V' or s[0] == 'I' and s[1] == 'X'
            # X can be placed before L (50) and C (100) to make 40 and 90. 
            or s[0] == 'X' and s[1] == 'L' or s[0] == 'X' and s[1] == 'C'
            # C can be placed before D (500) and M (1000) to make 400 and 900.
            or s[0] == 'C' and s[1] == 'D' or s[0] == 'C' and s[1] == 'M'
            ):
            # does'nt repeat the same index of the s. first 2 indexes are skipped
            val += romans[s[1]] - romans[s[0]]
            print(f"Block one: {val} s values {s[0]}, {s[1]}")
            s = s[2:]
            print('block one s: ', s)
        else:
            # does'nt repeat the same index of the s. first 1 index are skipped
            val += romans[s[0]]
            print(f"Block two: {val} s values {s[0]}")
            s = s[1:]
            print('block two s: ', s)
    return val

romanToInt(s="MCMXCIV")
```

Output: [1,1,2,3,4,4]
Example 2:

Input: list1 = [], list2 = []
Output: []
Example 3:

Input: list1 = [], list2 = [0]
Output: [0]
 

Constraints:

The number of nodes in both lists is in the range [0, 50].
-100 <= Node.val <= 100
Both list1 and list2 are sorted in non-decreasing order.

```python
from typing import Optional, List

class ListNode:
    def __init__(self, val = 0, next=None):
        self.val = val,
        self.next = next

class Solution:
    def merge_lists(self, list_1: Optional[ListNode], list_2: Optional[ListNode]) -> ListNode:
        if not list_1:
            return list_2

        if not list_2:
            return list_1

        current = temporary = ListNode()


        while list_1 and list_2:
            if list_1.val < list_2.val:
                current.next = list_1
                list_1 = list_1.next
            else:
                current.next = list_2
                list_2 = list_2.next
            current = current.next

            if list_1:
                current.next = list_1
            if list_2:
                current.next = list_2

            return temporary.next

            
        
  

a = Solution()
list1 = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5, None)))))
# ------------------------------------------------------------------------------------------
list2 = ListNode(1, ListNode(2, ListNode(3, None)))

print(a.merge_lists(list_1=list1, list_2=list2))
```

```python
def longestCommonPrefix(strs: List[str]) -> str:
    commonPrefix = ""
    for index, item in enumerate(strs):
        if index == 0:
            commonPrefix = item
        if not commonPrefix:
            break
        if len(commonPrefix) > len(item):
            commonPrefix = commonPrefix[:len(item)]
        for i, letter in enumerate(item):
            if len(commonPrefix) <= i:
                break
            if commonPrefix[i] != letter:
                commonPrefix = commonPrefix[:i]

    return commonPrefix


longestCommonPrefix(["aaa", "a", "aaa"])
```

```python
"""
find the difference between two 
"""
class Solution:
    def difference(self, s, t):
        for i in range(len(t)):
            if t[i] not in s:
                return t[i]
            if t.count(t[i])>s.count(t[i]):
                return t[i]

sol = Solution()
sol.difference(
    s='abc',
    t='abcd'
)
```

```python
"""
345. reverse the vowels in the given string.

"""


class Solution:
    def reverseVowels(self, s: str) -> str:
        vowel_data = {
            "a": "a",
            "A": "A",
            "e": "e",
            "E": "E",
            "i": "i",
            "I": "I",
            "o": "o",
            "O": "O",
            "u": "u",
            "U": "U",
        }
        temp = []
        vowels = []
        for i in s:
            if i in vowel_data:
                # if there is a vowel in the string append "_" to the temp list.
                temp.append("_")

                vowels.append(i)
            else:
                temp.append(i)

# using stack to reverse vowels pop and append
        # print(temp)
        # print(vowels)

        for i in temp:
            # check for _ in the temp list then replace it with the last value in the vowels list.
            if i == "_":
                # Append the last vowel based on the index of i
                temp[temp.index(i)] = vowels.pop()
        # print(temp)
        return "".join(temp)

    def reverseVowels2(self, s: str) -> str:
        vowel_data = {
            "a": "a",
            "A": "A",
            "e": "e",
            "E": "E",
            "i": "i",
            "I": "I",
            "o": "o",
            "O": "O",
            "u": "u",
            "U": "U",
        }
        s = list(s)
        i = 0
        j = len(s) - 1

        while i < j:
            if s[i] not in vowel_data:
                i += 1
                print(s)
            elif s[j] not in vowel_data:
                j -= 1
                print(s)
            else:
                s[i], s[j] = s[j], s[i]
                i += 1
                j -= 1
                print(s)
            return "".join(s)


sol = Solution()
# sol.reverseVowels(s="aA")
sol.reverseVowels2(s="qwerqwerqwrtyuiopfghjkl")
```
