```python
# # If Else and For loops.
df.insert(
    loc = 2,
    column = 'Price_Class',
    value = " "
)
df.head()

```

```python
for i in range(
    0, #start
    len(df['Price']), # stop
    1 # step
    ):

    # print(df['Price'][i])
    if df['Price'][i] <= 8000:
        df['Price_Class'][i] = 'Low'
    elif df['Price'][i] >= 11000:
        df['Price_Class'][i] = 'High'
    else:
        df['Price_Class'][i] = 'Medium'
```

```python
df.head()
```

```python
# using while loop

rows = 0
while rows < len(df['Price']):
    # print(df['Price'][rows])
    if df['Price'][rows] <=  8450:
        df['Price_Class'][rows] = 'Low'
    elif df['Price'][rows] >= 14000:
        df['Price_Class'][rows] = 'High'
    else: 
        df['Price_Class'][rows] = 'Medium'
    rows = rows + 1

```

```python
df.head()
```

```python
# count of unique vals in a column
df['Price_Class'].value_counts()
```

```python
# doing the same with a function.
def price_class(x):
    if x <= 8000:
        return 'Low'
    elif x >= 11000:
        return 'High'
    else:
        return 'Medium'



df['Price_Class'] = df['Price'].apply(price_class)
df['Price_Class'].value_counts()
```

```python
# taking in multiple inputs in a function and returning multiple vals.
def converter(
    val1, val2
):
    val_converted = val1 / 12 # total number of age in months
    ratio = val2 / val1 # km/month
    return [val_converted, ratio]
```

```python
df['age_converted'], df['km_per_month'] = converter(
    df['Age'], df['KM']
)
```
