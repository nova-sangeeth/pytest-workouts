```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    'Datasets/Toyota.csv',
    index_col=0,
    na_values = ['??', '????']
)
df.head()
```

```python
df.info()
```

```python
df['Doors'].replace('three', 3 ,inplace=True)
df['Doors'].replace('four', 4 ,inplace=True)
df['Doors'].replace('five', 5 ,inplace=True)
```

```python
# TWO WAY TABLES
# Using Cross tab to get a tabulation of one or more factors.

pd.crosstab(
    index=df['FuelType'], # col that needs to processed
    columns='count', # Create a new col and add the processed vals
    dropna=True, # Drop null vals.
)
```

```python
# Join Probability (Two Way Table)

pd.crosstab(
    index=df['Automatic'], # col that needs to processed
    columns=df['FuelType'],
    normalize=True
)
```

```python
# Marginal Probability

pd.crosstab(
    index = df['Automatic'],
    columns = df['FuelType'],
    margins = True,
    dropna=True,
    normalize=True
)
```

```python
# Conditional Probability using Two way table
pd.crosstab(
    index=df['Automatic'],
    columns=df['FuelType'],
    margins=True,
    # normalize='index'
    normalize='columns'
)
```
