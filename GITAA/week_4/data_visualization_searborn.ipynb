```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    'Datasets/Toyota.csv',
    index_col=0,
    na_values=['NA', '??', '????'],
    )
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(
    axis=0,
    inplace=True,
)

df.shape
```

# Scatter plot using seaborn.

```python
sns.set(
    style='darkgrid',
)

sns.regplot(
    x=df['Age'],
    y=df['Price'],
    fit_reg=False, # No regression line
    marker='*' # Changes the marker to a star
)
```

* scatter plot of price vs age by fuel type.


```python
#  lm plot combines facetgrid and regplot
sns.lmplot(
    x='Age',
    y='Price',
    data=df,
    fit_reg=True,
    hue='FuelType',
    legend=True,
    palette='Set1', # one of the color palettes
)
```

# sns.distplot Histogram.

```python
sns.distplot(
    df['Age'],
    kde=False, # No kernel density estimation
    bins=20,
    )

# sns.distplot(
#     df['Age'],
#     bins=5
# )
```

# bar plot

```python
sns.countplot(
    x='FuelType',
    data=df,
)
```

```python
# GROUPED BAR PLOT. (similar to cross tab using pandas.)
sns.countplot(
    x='FuelType',
    data=df,
    hue='Automatic'
)
```

# Box and whiskers plot for numerical variables.


five number summary includes.

* mininum
* maximum
* 25% first quantile
* 50% second quantile
* 75% third quantile



```python
sns.boxplot(
    y=df['Price']
)
# finds the median values and the quantile values of the price.
# the horizontal lines in the boxplot are the whiskers
#  There are two types of whiskers: the upper and the lower.
```

```python
# BOX AND WHISKERS PLOT.

# The middle line is the representation of the median in each box.
sns.boxplot(
    y=df['Price'],
    x=df['FuelType'],
    hue=df['Automatic']
)
```

* Box-Whiskers plot and Histogram in the same window.
* split the window in 2 parts.

```python
f, (ax_box, ax_hist) = plt.subplots(
    2, 
    gridspec_kw={
        "height_ratios": (.15, .85)
    }
)


#  create two plots.
sns.boxplot(
    df['Price'], ax=ax_box,
)
sns.distplot(
    df['Price'], ax=ax_hist,
    kde=True

)
```

* Pairwise plots
* User to plot pairwise relationships in a dataset.


```python
sns.pairplot(
    df, 
    kind='scatter',
    hue='Automatic'
)
plt.show()
```
